```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cpu = pd.read_csv('CPU_UserBenchmarks.csv')

gpu = pd.read_csv('GPU_UserBenchmarks.csv')

ssd = pd.read_csv('SSD_UserBenchmarks.csv')

hdd = pd.read_csv('HDD_UserBenchmarks.csv')

ram = pd.read_csv('RAM_UserBenchmarks.csv')


usb = pd.read_csv('USB_UserBenchmarks.csv')
usb

frames = [cpu, gpu, ssd, hdd, ram, usb]
df = pd.concat(frames)
df

dums = pd.get_dummies(df)
dums = dums.iloc[:, :9]
dums
```
